```python
import numpy as np
from sklearn.datasets import make_moons
from my_model import quantum_model as q_model
```

```python
weight_seed = 42
model = q_model(n_circuits = 2, rng_seed = weight_seed)
```

```python
print(model)
```

```python
test_size = 15
split_seed = 42
noise_value = 15
sample_size = 300
X, Y = make_moons(n_samples=sample_size, noise=noise_value/100, random_state=split_seed)
data_tuple = model.scale_and_split(X, Y, test_size = test_size/100, random_state = split_seed)
```

```python
n_epochs = 3
model.fit(data_tuple, alpha = 0.1, n_epochs = n_epochs, display = True)
#model.quick_fit(data_tuple, alpha = 0.1, n_epochs = n_epochs, display = True)
```

```python
fig = model.plotly_scatter(X[:,0], X[:,1], Y, title = "Data Visualization")
model.plot_list.append(fig)
fig.show()
```
